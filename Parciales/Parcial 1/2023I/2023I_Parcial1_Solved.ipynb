# 1. Contador de pares e impares

Escriba una función llamada `conteo(lst)` que reciba una lista de números y cuente la cantidad de números pares e impares. Por ejemplo, si la lista es:

> lst = [1, 3, 6, 7, 9, 10]

Debe imprimir

> Cantidad de números pares: 2.
>
> Cantidad de números impares: 4.

```python
def cont_par_impar(lista):
    
    pares = 0
    impares = 0
    
    for num in lista:
        
        if type(num) == int or type(num) == float:
            if num % 2 == 0:
                pares = pares + 1
            else:
                impares = impares + 1
        else:
            print("El elemento -{0}- no es de tipo numérico.".format(num))
            return # Stops the loop
        
    print("Cantidad de números pares: {0} \nCantidad de números impares: {1}".format(pares, impares))
```

```python
# Testing
lista = [1, 3, 6, 7, 9, 10]

cont_par_impar(lista)
```

# 3. Conteo de palabras iguales en una lista

Escriba una función `rep_words(lst, word)` que reciba una lista de palabras y una palabra específica y encuentre cuántas veces esa palabra se repite en la lista. Por ejemplo, si la lista es:

> `lst = ["hola", "HOLA", "hellow", "HoLaA", "HoLa"]`

Tenga los siguientes resultados:

> `rep_words(lst, "hola") = 3`
>
> `rep_words(lst, "hellow") = 1`
>
> `rep_words(lst, "mojarra") = 0`

```python
def rep_word(lst, word):

    word = word.upper()
    conteo = 0

    for element in lst:
        element = element.upper()

        if element == word:
            conteo += 1

    return conteo
```

```python
lst = ["hola", "HOLA", "hellow", "HoLaA", "HoLa"]
word = "mojarra"

rep_word(lst, word)
```

# 4. Divisores entre dos números

Escriba una función llamada `divisores(n, k)` que imprima todos los números que sean divisores a la vez de un número `n` y un número `k`. Por ejemplo,

> `divisores(24, 12)`

> `[out] 1, 2, 3, 4, 6, 12`

```python
def divisores(n, k):
    div = []

    if n >= k:
        max_number = n
    else:
        max_number = k

    for num in range(1, max_number + 1):
        if (n % num == 0) & (k % num == 0):
            div.append(num)

    return div
```

```python
for n in [3, 8, 15, 48]:
    for k in [3, 4, 5, 8, 12, 24]:
        print(n, k, "\t", divisores(n, k))
```

# 5. Números amigables

Los números amigables son un par de números cuyos divisores (excluyéndose a sí mismo) sumados obtienen el otro número. Por ejemplo, 220 y 284 son números amigables, ya que los divisores de 220 son 

$$ Divisores220 : 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 $$

Ahora, si se suman esos divisores, se obtiene que

$$ \sum Divisores220 = 1 + 2 + 4 + 5 + 10 + 11 + 20 + 22 + 44 + 55 + 110 = 284 $$

Veamos para 284

$$ Divisores284 : 1, 2, 4, 71, 142 $$

$$ \sum Divisores284 = 1 + 2 + 4 + 71 + 142 = 220 $$

Otros pares de números amigos son 1184-1210, 2620-2924, 5020-5564, entre otros.

Teniendo en cuenta lo anterior, construya una función `Amix_Number(x, y)` que reciba dos números (x e y) y arroje si son números amigables.


```python
def amix_number(x, y):

    x_div = []
    y_div = []

    for i in range(1, x):
        if x % i == 0:
            x_div.append(i)

    for i in range(1, y):
        if y % i == 0:
            y_div.append(i)

    #print(x_div)
    #print(y_div)
    #print(sum(x_div))
    #print(sum(y_div))

    if (sum(x_div) == y) & (sum(y_div) == x):
        return "The numbers {0} and {1} are amical numbers.".format(x, y)
    else:
        return "The numbers {0} and {1} are not amical numbers.".format(x, y)
```

```python
lst = [[220, 284], [1184, 1210], [2620, 2924], [1, 2]]

for element in lst:
    print(element)
    print(amix_number(element[0], element[1]))
```
