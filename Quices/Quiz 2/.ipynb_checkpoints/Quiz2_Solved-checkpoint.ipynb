Responda a las siguientes preguntas:
* ¿Cuál es la ciudad de origen que registra mayor número de vuelos atrasados?
* ¿Cuál es la ciudad de origen que registra mayor número de vuelos cancelados?
* ¿Cuál es la ciudad de destino que registra mayor número de vuelos atrasados?
* ¿Cuál es la ciudad de destino que registra mayor número de vuelos cancelados?
* ¿Cuál es la combinación de ciudad origen y destino con mayor número de vuelos atrasados?
* ¿Cuál es la combinación de ciudad origen y destino con mayor número de vuelos cancelados?
* ¿Aparece algún nombre en particular repetitivamente?

```python
table_delay = data[["OriginCityName", "Flights"]][data.DepDelayMinutes > 0]\
    .groupby("OriginCityName").count()\
    .sort_values("Flights", ascending = False)\
    .reset_index()

table_cancelled = data[["OriginCityName", "Flights"]][data.Cancelled == 1]\
    .groupby("OriginCityName").count()\
    .sort_values("Flights", ascending = False)\
    .reset_index()


print("La ciudad de origen con mayor número de vuelos atrasados es {0} con {1} vuelos retrasados."\
          .format(table_delay["OriginCityName"][0], table_delay["Flights"][0]))
print("La ciudad de origen con mayor número de vuelos cancelados es {0} con {1} vuelos cancelados"\
          .format(table_cancelled["OriginCityName"][0], table_cancelled["Flights"][0]))
```

```python
table_delay = data[["DestCityName", "Flights"]][data.DepDelayMinutes > 0]\
    .groupby("DestCityName").count()\
    .sort_values("Flights", ascending = False)\
    .reset_index()

table_cancelled = data[["DestCityName", "Flights"]][data.Cancelled == 1]\
    .groupby("DestCityName").count()\
    .sort_values("Flights", ascending = False)\
    .reset_index()


print("La ciudad de destino con mayor número de vuelos atrasados es {0} con {1} vuelos retrasados."\
          .format(table_delay["DestCityName"][0], table_delay["Flights"][0]))
print("La ciudad de destino con mayor número de vuelos cancelados es {0} con {1} vuelos cancelados"\
          .format(table_cancelled["DestCityName"][0], table_cancelled["Flights"][0]))
```

```python
table_delay = data[["OriginCityName", "DestCityName", "Flights"]][data.DepDelayMinutes > 0]\
    .groupby(["OriginCityName", "DestCityName"]).count()\
    .sort_values("Flights", ascending = False)\
    .reset_index()

table_cancelled = data[["OriginCityName", "DestCityName", "Flights"]][data.Cancelled == 1]\
    .groupby(["OriginCityName", "DestCityName"]).count()\
    .sort_values("Flights", ascending = False)\
    .reset_index()


print("La combinación de ciudades con mayor número de vuelos atrasados son {0} como origen y {1} como destino con {2} vuelos retrasados."\
          .format(table_delay["OriginCityName"][0], table_delay["DestCityName"][0], table_delay["Flights"][0]))
print("La combinación de ciudades con mayor número de vuelos cancelados son {0} como origen y {1} como destino con {2} vuelos cancelados."\
          .format(table_cancelled["OriginCityName"][0], table_cancelled["DestCityName"][0], table_cancelled["Flights"][0]))
```

```python
import seaborn as sns

# Estableciendo tamaño de la gráfica
plt.figure(figsize=(15, 45), dpi = 80) 

# Variables a graficar
vars_to_plot = data["Operating_Airline "].unique()

# Se hace un ciclo
for i, var in enumerate(vars_to_plot):
    #print(i, var) 
    plt.subplot(11, 2, i+1) # Se ubica cada gráfica en una subgráfica
    sns.scatterplot(x = 'ActualElapsedTime', y = 'DepDelayMinutes',\
                    hue = 'Operating_Airline ', \
                    data = data[data["Operating_Airline "] == var])
    title_string = "Histogram of " + var # Se crea una variable del título
    plt.ylabel("Retraso del vuelo (en minutos)"); plt.xlabel("")
    plt.title("Gráfico de distribución del Retraso del Vuelo vs. Distancia")
    
```

```python
# Se obtienen los barrios
airlines = data["Operating_Airline "].unique()

# Se crea un diccionario para tener las correlaciones por barrios
cvec = dict()

# Se agrupa por barrios y se calcula la correlación
for x in airlines:
    temp = data[data['Operating_Airline '] == x]
    cvec[x] = temp.corr()['DepDelayMinutes']['ActualElapsedTime']
    
# Se guardan los resultados
res = list(cvec.values())

# Se borran aquellos que tienen missings
res = [x for x in res if str(x) != 'nan']

# Se organizan los datos
res.sort()

# Se hace un histograma
plt.hist(res, bins=50)
plt.ylabel('Correlation')
plt.show()

# Se obtiene la correlación promedio entre barrios
print('Average correlation: ', sum(res)/len(res))
```

## 4.2. Tiempo de retraso de vuelo y la fecha del vuelo

Realice un gráfico de serie de tiempo (puede ser un gráfico de barras o uno de líneas) en el que se encuentre cuál fue el retraso promedio por fecha de vuelo:
* ¿Qué puede observar?
* ¿Existe un comportamiento que responde a lógicas que se repiten en el tiempo? Es decir, ¿es periódico?

```python
# Setting the Date in Datetime format
data["FlightDate"] = pd.to_datetime(data.FlightDate)
```

```python
# Making a temporal dataframe
df_temp = data[["FlightDate", "DepDelayMinutes"]].groupby("FlightDate").mean()

# Plotting
plt.figure(figsize=(15, 8), dpi = 80) 
plt.plot(df_temp)
plt.title("Tiempo de demora promedio por día")
plt.xlabel("Fecha");plt.ylabel("Tiempo de retraso (en minutos)")
plt.show()
```

Ahora realice un gráfico de barras con el retraso promedio de los vuelos por días de la semana (Lunes-Domingo):
* ¿Hay días que tienen un promedio de retraso mayor que otros? ¿Cuáles?

```python
# Generating a copy of the dataframe
df_temp = data.copy()

# Generating a temporal for the mean of delays by week day
df_temp["WeekDay"] = [day.strftime('%A') for day in df_temp["FlightDate"]]
df_temp = df_temp[["WeekDay", "DepDelayMinutes"]].groupby("WeekDay").mean().reset_index()

# Changing the type of weekday to have it ordered
df_temp["WeekDay"] = pd.Categorical(df_temp["WeekDay"],  \
                                    ordered = True,\
                                    categories = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
```

```python
# Plotting
plt.figure(figsize = (15, 8))
sns.barplot(x = "WeekDay",
            y = "DepDelayMinutes",
            data = df_temp,
            color='#69b3a2')
plt.title("Gráfico de barras del retraso promedio por día de la semana")
plt.xlabel("Día de la Semana");plt.ylabel("Retraso promedio (en minutos)")
plt.show()
```
