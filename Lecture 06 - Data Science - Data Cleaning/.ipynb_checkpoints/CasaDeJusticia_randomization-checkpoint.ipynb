```python
# Importing library
import random
import pandas as pd

# Setting the seed
random.seed(7)

# Generating the lists to be random
days = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
range_days = ["7:30 - 9:00", "9:00 - 12:00", "12:00 - 2:30", "3:00 - 4:00"]
casas = ["Ciudad Bolívar", "Bosa y Kennedy", "Usme", "Suba Jardín", "Casa Móvil"]
caso = ["Arriendo - Ocupación Indebida", "Arriendo - Otra Cosa", "VIF - LGBT Propio", "VIF - LGBT Tercero", "VIF - Adulto Mayor"]

# Generating the samples
random_week  = random.sample(list(range(1, 5)) + random.sample(range(1, 5), 1), 5)
random_day   = random.sample(days, 5)
random_range = random.sample(range_days + random.sample(range_days, 1), 5)
random_casa  = random.sample(casas, 5)
random_caso  = random.sample(caso, 5)

# Generating the dataframe
i = 0
for casa in random_casa:
    
    if i == 0:
        random_week  = random.sample(list(range(1, 5)) + random.sample(range(1, 5), 1), 5)
        random_day   = random.sample(days, 5)
        random_range = random.sample(range_days + random.sample(range_days, 1), 5)
        random_casa  = random.sample(casas, 5)
        random_caso  = random.sample(caso, 5)
    
        res = pd.DataFrame({"Casa": [casa]*5, "Caso": random_caso, "Semana": random_week, "Dia": random_day, "RangoDia": random_range})
        i += 1
    
    else:
        random_week  = random.sample(list(range(1, 5)) + random.sample(range(1, 5), 1), 5)
        random_day   = random.sample(days, 5)
        random_range = random.sample(range_days + random.sample(range_days, 1), 5)
        random_casa  = random.sample(casas, 5)
        random_caso  = random.sample(caso, 5)
    
        df_temp = pd.DataFrame({"Casa": [casa]*5, "Caso": random_caso, "Semana": random_week, "Dia": random_day, "RangoDia": random_range})
        res = pd.concat([res, df_temp])
        

# Getting a shorter version of the dataframe
results = pd.DataFrame({"Semana": res["Semana"], "Dia": res["Dia"],\
             "Detalles": res["Casa"] + ", " + res["Caso"] + ", " + res["RangoDia"]})

res.to_excel
```
